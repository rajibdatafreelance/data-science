### FEATURES

The various features of the cleaned dataset are explained below:

    1) Airline: The name of the airline company is stored in the airline column. It is a categorical feature having 6 different airlines. 
    2) Flight: Flight stores information regarding the plane's flight code. It is a categorical feature. 
    3) Source City: City from which the flight takes off. It is a categorical feature having 6 unique cities. 
    4) Departure Time: This is a derived categorical feature obtained created by grouping time periods into bins. It stores information about the departure time and have 6 unique time labels. 
    5) Stops: A categorical feature with 3 distinct values that stores the number of stops between the source and destination cities. 
    6) Arrival Time: This is a derived categorical feature created by grouping time intervals into bins. It has six distinct time labels and keeps information about the arrival time. 
    7) Destination City: City where the flight will land. It is a categorical feature having 6 unique cities. 
    8) Class: A categorical feature that contains information on seat class; it has two distinct values: Business and Economy. 
    9) Duration: A continuous feature that displays the overall amount of time it takes to travel between cities in hours. 
    10)Days Left: This is a derived characteristic that is calculated by subtracting the trip date by the booking date. 
    11) Price: Target variable stores information of the ticket price.


```python
# importing all libreries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# importing the dataset
df = pd.read_excel("flight_price.xlsx")
df.head()
```

```python
# Checking the summary of data
df.info()
```

```python
# Checking null values
df.isnull().sum()
```

```python
# Feature Engineering on column "date_of_journey"
df["Date"]= df["Date_of_Journey"].str.split("/").str[0]
df["Month"]= df["Date_of_Journey"].str.split("/").str[1]
df["Year"]= df["Date_of_Journey"].str.split("/").str[2]
df.head()
```

```python
# Converting new columns to int
df["Date"]=df["Date"].astype(int)
df["Month"]=df["Month"].astype(int)
df["Year"]=df["Year"].astype(int)
df.info()
```

```python
#Now we can drop Date_of_journey column
df.drop("Date_of_Journey", axis=1, inplace=True)
df.head()
```

```python
# Feature Engineering on Arrival time
df["Arrival_Time"] = df["Arrival_Time"].apply(lambda x:x.split(" ")[0])
df["Arrival_hour"]= df["Arrival_Time"].str.split(":").str[0]
df["Arrival_minute"]= df["Arrival_Time"].str.split(":").str[1]
```

```python
# Converting output column into int
df["Arrival_hour"]= df["Arrival_hour"].astype(int)
df["Arrival_minute"]= df["Arrival_minute"].astype(int)
```

```python
df.info()
```

```python
# Now drop Arrival time
df.drop("Arrival_Time", axis=1, inplace=True)
```

```python
df.head()
```

```python
# Feature Engineering on Departure time
df['Departure_hour']=df['Dep_Time'].str.split(':').str[0]
df['Departure_min']=df['Dep_Time'].str.split(':').str[1]
```

```python
# Converting new columns into int
df["Departure_hour"]=df["Departure_hour"].astype(int)
df["Departure_min"]=df["Departure_min"].astype(int)
```

```python
df.info()
```

```python
# Drop Dep_time column
df.drop("Dep_Time", axis=1, inplace=True)
df.head(3)
```

```python
# Feature Engineering on total_stops
df["Total_Stops"].unique()
```

```python
df["Total_Stops"].value_counts()
```

```python
# Replacing the null value with the mode value and other values with numeric values
df["Total_Stops"]= df["Total_Stops"].map({ "1 stop": 1, "non-stop" : 0, "2 stops" : 2, "3 stops": 3 , "4 stops": 4, np.nan:1})
```

```python
df.head(3)
```

```python
# Drop "Route" column as we have source and destination columns
df.drop("Route", axis=1, inplace = True)
```

```python
# Feature Engineering on "Duration" column- only extract the hour part
df["Duration_hour"]=df["Duration"].str.split(" ").str[0].str.split("h").str[0]
df["Duration_minute"]=df["Duration"].str.split(" ").str[0].str.split("h").str[0]
```

```python
# drop Duration column
df.drop("Duration", axis=1, inplace=True)
```

```python
df.head()
```

```python
# Checking unique values of Airline
df["Airline"].unique()
```

```python
# Checking unique values of Source
df["Source"].unique()
```

```python
# Checking unique values of Destination
df["Destination"].unique()
```

```python
# Checking unique values of "Additional Info"
df["Additional_Info"].unique()
```

```python
# Applying one hot encoding to above categorical features
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
encoder.fit_transform(df[["Source","Destination","Additional_Info", "Airline"]]).toarray()
```

```python
# Converting into dataframe 
df1 = pd.DataFrame(encoder.fit_transform(df[['Airline','Source','Destination']]).toarray(),columns=encoder.get_feature_names_out())
df1
```

```python
# concating the new dataframe with original dataframe
final_df = pd.concat([df,df1], axis=1)
```

```python
# checking the final dataframe
final_df.info()
```

```python
#Drop the old columns
final_df.drop(["Airline", "Source", "Destination"], axis=1, inplace=True)
```
